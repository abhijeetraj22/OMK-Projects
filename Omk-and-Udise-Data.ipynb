```python
import pandas as pd
from google.colab import files

def match_spreadsheets():
    # Upload file
    uploaded = files.upload()
    excel_file = list(uploaded.keys())[0]

    # Define sheet names and columns
    sheet1 = 'OMK'
    sheet2 = 'UDISE'
    col_name = 'Name'
    col_father = 'Father Name'
    col_mother = 'Mother Name'

    # Load the spreadsheets
    df1 = pd.read_excel(excel_file, sheet_name=sheet1)
    df2 = pd.read_excel(excel_file, sheet_name=sheet2)

    # Strip spaces from column names
    df1.columns = df1.columns.str.strip()
    df2.columns = df2.columns.str.strip()

    # Check if all required columns are present
    for col in [col_name, col_father, col_mother]:
        if col not in df1.columns:
            print(f"Missing column in {sheet1}: {col}")
            return
        if col not in df2.columns:
            print(f"Missing column in {sheet2}: {col}")
            return

    # Perform row-wise match on Name, Father Name, and Mother Name
    df2['Match Status'] = df2.apply(
        lambda row: (
            (df1[col_name] == row[col_name]) &
            (df1[col_father] == row[col_father]) &
            (df1[col_mother] == row[col_mother])
        ).any(),
        axis=1
    )

    matched_data = df2[df2['Match Status']]
    unmatched_data = df2[~df2['Match Status']]

    # Display matched and unmatched data
    from IPython.display import display
    print("✅ Matched Records:")
    display(matched_data)

    print("❌ Unmatched Records:")
    display(unmatched_data)

    # Show match statistics
    matched_count = matched_data.shape[0]
    total_count = len(df2)
    unmatched_count = total_count - matched_count

    print("\n📊 Match Statistics:")
    print(f"Total records in {sheet2}: {total_count}")
    print(f"Matched records: {matched_count}")
    print(f"Unmatched records: {unmatched_count}")

# Run the function in Colab
match_spreadsheets()
```

```python
import pandas as pd
from google.colab import files

def match_spreadsheets():
    # Upload the Excel file
    uploaded = files.upload()
    excel_file = list(uploaded.keys())[0]

    # Sheet names
    sheet1 = 'OMK'
    sheet2 = 'UDISE'

    # Columns to match
    col_name = 'Name'
    col_father = 'Father Name'
    col_mother = 'Mother Name'
    sid_col = 'SID'
    pen_col = 'Student PEN'

    # Load the sheets
    df1 = pd.read_excel(excel_file, sheet_name=sheet1)
    df2 = pd.read_excel(excel_file, sheet_name=sheet2)

    # Strip column names
    df1.columns = df1.columns.str.strip()
    df2.columns = df2.columns.str.strip()

    # Check required columns
    for col in [col_name, col_father, col_mother]:
        if col not in df1.columns or col not in df2.columns:
            print(f"❌ Missing '{col}' in one of the sheets.")
            return
    if sid_col not in df1.columns:
        print(f"❌ Missing '{sid_col}' in Sheet1.")
        return
    if pen_col not in df2.columns:
        print(f"❌ Missing '{pen_col}' in Sheet2.")
        return

    # Create match key
    def create_key(df):
        return (
            df[col_name].astype(str).str.strip().str.lower() + "|" +
            df[col_father].astype(str).str.strip().str.lower() + "|" +
            df[col_mother].astype(str).str.strip().str.lower()
        )

    df1['match_key'] = create_key(df1)
    df2['match_key'] = create_key(df2)

    # Merge SID from Sheet1 into Sheet2
    merged = df2.merge(df1[['match_key', sid_col]], on='match_key', how='left')

    # Mark match status
    merged['Match Status'] = merged[sid_col].notna()
    merged[sid_col] = merged[sid_col].fillna('Not Found')

    # Select final columns
    output_cols = [col_name, col_father, col_mother, pen_col, sid_col, 'Match Status']

    # Show results
    from IPython.display import display
    print("✅ Matched Records:")
    display(merged[merged['Match Status']][output_cols])

    print("\n❌ Unmatched Records:")
    display(merged[~merged['Match Status']][output_cols])

    # Stats
    print("\n📊 Summary:")
    print(f"Total: {len(df2)}")
    print(f"Matched: {merged['Match Status'].sum()}")
    print(f"Unmatched: {(~merged['Match Status']).sum()}")

# Run
match_spreadsheets()
```

```python
import pandas as pd

def match_spreadsheets(file_path):
    # Sheet names
    sheet1 = 'OMK'
    sheet2 = 'UDISE'

    # Columns to match
    col_name = 'Name'
    col_father = 'Father Name'
    col_mother = 'Mother Name'
    sid_col = 'SID'
    pen_col = 'Student PEN'

    # Load both sheets
    df1 = pd.read_excel(file_path, sheet_name=sheet1)
    df2 = pd.read_excel(file_path, sheet_name=sheet2)

    # Clean column names
    df1.columns = df1.columns.str.strip()
    df2.columns = df2.columns.str.strip()

    # Validate columns
    for col in [col_name, col_father, col_mother]:
        if col not in df1.columns or col not in df2.columns:
            print(f"❌ Missing '{col}' in one of the sheets.")
            return
    if sid_col not in df1.columns:
        print(f"❌ Missing '{sid_col}' in Sheet1.")
        return
    if pen_col not in df2.columns:
        print(f"❌ Missing '{pen_col}' in Sheet2.")
        return

    # Create a matching key
    def create_key(df):
        return (
            df[col_name].astype(str).str.strip().str.lower() + "|" +
            df[col_father].astype(str).str.strip().str.lower() + "|" +
            df[col_mother].astype(str).str.strip().str.lower()
        )

    df1['match_key'] = create_key(df1)
    df2['match_key'] = create_key(df2)

    # Merge on match_key
    merged = df2.merge(df1[['match_key', sid_col]], on='match_key', how='left')

    # Convert SID to integer (if possible), else fill NA
    merged[sid_col] = pd.to_numeric(merged[sid_col], errors='coerce').fillna(0).astype(int)
    merged[sid_col] = merged[sid_col].replace(0, 'Not Found')

    # Mark match status
    merged['Match Status'] = merged[sid_col] != 'Not Found'

    # Select final columns
    output_cols = [col_name, col_father, col_mother, pen_col, sid_col, 'Match Status']

    # Show results
    matched = merged[merged['Match Status']][output_cols]
    unmatched = merged[~merged['Match Status']][output_cols]

    print("✅ Matched Records:")
    print(matched.to_string(index=False))

    print("\n❌ Unmatched Records:")
    print(unmatched.to_string(index=False))

    print("\n📊 Summary:")
    print(f"Total: {len(df2)}")
    print(f"Matched: {len(matched)}")
    print(f"Unmatched: {len(unmatched)}")

# 👇 Pass the path to the Excel file here
match_spreadsheets("MKS-UDise-2024-25 New.xlsx")
```

```python
import pandas as pd

def match_spreadsheets(file_path, output_file):
    # Sheet names
    sheet1 = 'OMK'
    sheet2 = 'UDISE'

    # Columns
    col_name = 'Name'
    col_father = 'Father Name'
    col_mother = 'Mother Name'
    sid_col = 'SID'
    pen_col = 'Student PEN'

    # Load sheets
    df1 = pd.read_excel(file_path, sheet_name=sheet1)
    df2 = pd.read_excel(file_path, sheet_name=sheet2)

    # Clean column names
    df1.columns = df1.columns.str.strip()
    df2.columns = df2.columns.str.strip()

    # Check required columns
    for col in [col_name, col_father, col_mother]:
        if col not in df1.columns or col not in df2.columns:
            print(f"❌ Missing '{col}' in one of the sheets.")
            return
    if sid_col not in df1.columns:
        print(f"❌ Missing '{sid_col}' in Sheet1.")
        return
    if pen_col not in df2.columns:
        print(f"❌ Missing '{pen_col}' in Sheet2.")
        return

    # Create a match key
    def make_key(df):
        return (
            df[col_name].astype(str).str.strip().str.lower() + "|" +
            df[col_father].astype(str).str.strip().str.lower() + "|" +
            df[col_mother].astype(str).str.strip().str.lower()
        )

    df1['match_key'] = make_key(df1)
    df2['match_key'] = make_key(df2)

    # Merge on match_key to get SID
    merged = df2.merge(df1[['match_key', sid_col]], on='match_key', how='left')

    # Convert SID from float to int, or mark as Not Found
    merged[sid_col] = pd.to_numeric(merged[sid_col], errors='coerce').fillna(0).astype(int)
    merged[sid_col] = merged[sid_col].replace(0, 'Not Found')

    # Match status
    merged['Match Status'] = merged[sid_col] != 'Not Found'

    # Final columns to keep
    output_cols = [col_name, col_father, col_mother, pen_col, sid_col, 'Match Status']

    matched = merged[merged['Match Status']][output_cols]
    unmatched = merged[~merged['Match Status']][output_cols]

    # Save to new Excel file
    with pd.ExcelWriter(output_file) as writer:
        matched.to_excel(writer, sheet_name='Matched', index=False)
        unmatched.to_excel(writer, sheet_name='Unmatched', index=False)

    print(f"✅ Results saved to: {output_file}")
    print(f"Matched: {len(matched)} | Unmatched: {len(unmatched)}")

# 🔽 Call the function with file path
match_spreadsheets(
    file_path="MKS-UDise-2024-25 New.xlsx",
    output_file="OMK FINAL Matched_Unmatched_Result.xlsx"
)
```
